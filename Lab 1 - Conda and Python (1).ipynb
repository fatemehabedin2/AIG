
# Introduction to Python Basics

Python is a popular programming language for data analysis and machine learning. In this section, we'll cover basic Python syntax and data types.

## Task
Write a Python script that takes a list of numbers and returns a list containing only the even numbers.

## Assessment Question
- Explain the difference between a list and a tuple in Python.


```python
# a list and a tuple are data collections. The difference is that list is mutatable means that you can change the items after creating the list so it could be used for dynamic data. However it's performance is lower and needs more memory,
# but tuple is immutable and you can't change it after creation so it can be used only for fixed data.

def even_num (nums: list) -> list:
    even_nums = []
    for num in nums:
        if num % 2 == 0:
            even_nums.append(num)
    return even_nums

numbers = [1, 2, 7, 3, 4, 5, 8, 6, 7, 8, 9, 10]
even_num (numbers)
```


# Working with NumPy

NumPy is a fundamental package for scientific computing with Python. It provides support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

## Task
Create a NumPy array and perform basic array operations (addition, multiplication).

## Assessment Question
- What are the advantages of using NumPy arrays over standard Python lists for numerical data?


```python
# NumPy arrays make operation much faster and memory efficient. it also supports vectorised operations (element-wise operation) without creating loops so code is simpler 
# we can also use numpy functions for operations on arrays, slicing and indexing. Use of multidimentional arrays is another benefit of arrays.

import numpy as np

array1 = np.array ([2, 6, 8, 10])
array2 = np.array ([20, 60, 80, 10])

sum_array = array1 + array2
print("sum of arrays = ", sum_array)

multiply_array = array1 * array2
print("product of arrays = ", multiply_array)
```

```python
# Scalar addition
scalar_addition = array1 + 5
print("Array scalar addition:", scalar_addition)
```

```python
# Scalar multiplication
scalar_multiplication = array1 * 3
print("Array scalar multiplication:", scalar_multiplication)
```


# Data Visualization with Matplotlib

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

## Task
Create a basic plot (line plot or bar chart) using data from a Pandas DataFrame.

## Assessment Question
- Why is data visualization important in data analysis?


```python
# Data visualization helps to understand complicated data faster, shows patterns, relationships, anomalies and outliers. 
# It improves story telling, communication and decision making.  


import matplotlib.pyplot as plt

columns = ['USERS', 'PEDESTRIANS', 'CARS']
grouped_by_location = ATM_df.groupby('LOCATION')[columns].sum()

grouped_by_location.plot(kind='bar', figsize=(10, 6))

plt.title("Total Counts by Location")
plt.xlabel("Location")
plt.ylabel("Total Count")
plt.legend()
plt.show()

```


# Introduction to SciPy

SciPy is a Python-based ecosystem of open-source software for mathematics, science, and engineering.

## Task
Demonstrate a simple mathematical operation using SciPy, like solving a linear equation.

## Assessment Question
- What is the role of SciPy in scientific computing with Python?


```python
# SciPy provides advanced mathematical functions that are not available in NumPy, for example linear algebra, integration, optimization , statistical analysis, and interpolation even signal and image processing . 
# also it has high performance algoruythms with high scalability. It also works well with other python libraries like Pandas, Scikit-learn and Matplotlib.

# we can use scipy to solve below equations
# x + y = 6              >>  [1, 1]
# 3x - y = 2             >>  [3, -1]
from scipy.linalg import solve          # SciPy Function solve(A, b), solves for x in the equation Ax = b where A is a 2D NumPy array representing the coefficient matrix. and b is a 1D NumPy array representing the constants.
import numpy as np
A = np.array([[1, 1],
              [3, -1]]) 
b = np.array([6, 2])

result = solve(A, b)
print(f"x, y = {result}")
```
